```python
import pandas as pd
import numpy as np
import os
import gc

import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib_venn import venn2, venn2_circles

import warnings
warnings.filterwarnings('ignore')
sns.set()
%matplotlib inline
```

```python
PATH_TO_DATA = ('D:/Py/DataFrames/Home_Credit_Default_Risk(KAGGLE)/')
```

---

```python
id_col = ['SK_ID_CURR']
id_cols = ['SK_ID_CURR', 'SK_ID_PREV']

app_train = pd.read_csv(os.path.join(PATH_TO_DATA, 'application_train.csv'), usecols=id_col)
app_test = pd.read_csv(os.path.join(PATH_TO_DATA, 'application_test.csv'), usecols=id_col)

bureau = pd.read_csv(os.path.join(PATH_TO_DATA, 'bureau.csv'), usecols=['SK_ID_CURR', 'SK_ID_BUREAU'])
bb = pd.read_csv(os.path.join(PATH_TO_DATA, 'bureau_balance.csv'), usecols=['SK_ID_BUREAU'])
pos_cash = pd.read_csv(os.path.join(PATH_TO_DATA, 'POS_CASH_balance.csv'), usecols=id_cols)
cc = pd.read_csv(os.path.join(PATH_TO_DATA, 'credit_card_balance.csv'), usecols=id_cols)
prev_app = pd.read_csv(os.path.join(PATH_TO_DATA, 'previous_application.csv'), usecols=id_cols)
inst = pd.read_csv(os.path.join(PATH_TO_DATA, 'installments_payments.csv'), usecols=id_cols)
```

```python
fig, axarr = plt.subplots(2, 2, figsize=(14, 7))

axarr[0, 0].set_title(f'Overlap between SK_ID_CURR in app_train and app_test')
venn2([
    set(app_train['SK_ID_CURR'].values), 
    set(app_test['SK_ID_CURR'].values)
], set_labels = ('train', 'test'), ax=axarr[0, 0])

axarr[0, 1].set_title(f'Overlap between SK_ID_BUREAU in bureau and bb')
venn2([
    set(bureau['SK_ID_BUREAU'].values), 
    set(bb['SK_ID_BUREAU'].values)
], set_labels = ('bureau', 'bb'), ax=axarr[0, 1])

axarr[1, 0].set_title(f'Overlap between SK_ID_CURR in bureau and app_train')
venn2([
    set(bureau['SK_ID_CURR'].values), 
    set(app_train['SK_ID_CURR'].values)
], set_labels = ('bureau', 'app_train'), ax=axarr[1, 0])

axarr[1, 1].set_title(f'Overlap between SK_ID_CURR in bureau and app_test')
venn2([
    set(bureau['SK_ID_CURR'].values), 
    set(app_test['SK_ID_CURR'].values)
], set_labels = ('bureau', 'app_test'), ax=axarr[1, 1]);
```
